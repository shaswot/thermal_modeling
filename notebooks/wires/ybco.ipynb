```python
# 2021solovyovYBCOonKaptonMaterialHighDensity
# https://www.osti.gov/servlets/purl/1764583

conductivity = 12 # W per sqm for 1 m long cable between 50K and 4K # overestimation

width = 1E-3 # line width
thickness = 100E-6 + 1E-6 #100um(kapton) + 1um(YBCO film)
area = width * thickness
# length = 460E-3 # 460 mm0

# krinner
l_50K = 20E-2 
l_4K = 29E-2 
l_Still = 25E-2 
l_CP = 17E-2 
l_MXC = 14E-2 
```

```python
PHL_YBCO_4K = conductivity * (area/l_4K)
print(f"{PHL_YBCO_4K=:.5e}")
```
