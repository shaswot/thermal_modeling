```python
import sys
import git
import pathlib

# Set up the PROJ_ROOT variable
PROJ_ROOT_PATH = pathlib.Path(git.Repo('.', search_parent_directories=True).working_tree_dir)
PROJ_ROOT =  str(PROJ_ROOT_PATH)
if PROJ_ROOT not in sys.path:
    sys.path.append(PROJ_ROOT)

# Explicitly add the current notebook's directory
CURRENT_DIR = str(pathlib.Path().absolute())
if CURRENT_DIR not in sys.path:
    sys.path.insert(0, CURRENT_DIR)
```

```python
import nbformat
from nbconvert.preprocessors import ExecutePreprocessor
import os
```

```python
# Set a timeout for cell execution in seconds
# Increase this if you have long-running cells (e.g., to 600 for 10 minutes)
timeout = 600 
```

```python
# fridges = ["xld1000sl","kide"]
fridges = ["kide"]
amps  = ["hemt", "hemt-lp", "hemt-ulp", "sisv1"]
wires = ["cu", "mn", "ybco"]
for fridge in fridges:
    for amp in amps:
        for wire in wires:
            exp = "ff-fc-delft-"+ amp + "_" + wire + "-" + fridge
            fname = PROJ_ROOT_PATH / "notebooks" / "experiments" / fridge / amp / exp / f"{exp}.ipynb"
    
            # Define the notebook file name
            notebook_filename = fname
    
            # Load the notebook
            with open(notebook_filename) as f:
                nb = nbformat.read(f, as_version=4)
    
            # Configure the ExecutePreprocessor
            ep = ExecutePreprocessor(timeout=timeout, kernel_name='python3')
    
            # print(f"Executing notebook: {notebook_filename}")
    
            try:
                # Execute the notebook
                # The 'resources' dictionary is required but can be empty
                ep.preprocess(nb, {'metadata': {'path': os.path.dirname(notebook_filename) or '.'}})
                print("Execution successful.")
            except Exception as e:
                print(f"Error during notebook execution: {e}")
                # You might want to handle errors differently, e.g., save the notebook up to the error
                # with open('executed_error_notebook.ipynb', 'w', encoding='utf-8') as f:
                #     nbformat.write(nb, f)
                # raise
            
            finally:
                # Save the executed notebook (even if an error occurred, to inspect outputs)
                with open(notebook_filename, 'w', encoding='utf-8') as f:
                    nbformat.write(nb, f)
                print(f"Executed notebook saved to {notebook_filename}")

```
