```python
att = 20 # attenuation at 4K for flux line
Z0 = 50 # characteristic impedance of flux line
```

```python
N = 10**(att/20) # V_in/V_out
R = Z0*(N-1)/(N+1) # Twin resistor value
Rs = 2*N*Z0/(N**2 - 1) # Shunt resistor value
```

```python
print(f"{N=}, {R=}, {Rs=}")
```

```python
Z = 50 # During AC operation i.e., filtering out high frequency thermal noise
R_eqv = R + (R+Z)**2/Rs + (R+Rs+Z)**2 * R/Rs**2 # the effective resistance w.r.t. output current
print(f"{R_eqv=:.2e}")
```

Theoretical Check: A 20 dB attenuator reduces power by a factor of 100 ($P_{in}  =100 \times P_{out}$). The total power dissipated is $P_{in} − P_{out} =99 P_{out}$.

Since $P_{out} = I_{out}^2 \times Z$, the dissipated power is $99\times I_{out}^2\times Z$.

With $Z =$ 50Ω, the theoretical effective resistance is $99\times50=4,950$Ω.

```python
Z = 0 # During DC operation, the load is effectively short circuited
R_eqv = R + (R+Z)**2/Rs + (R+Rs+Z)**2 * R/Rs**2 # the effective resistance w.r.t. output current
print(f"{R_eqv=:.2e}")
```

```python
R_eqv
```

```python
I_out = 0.4E-3 # Requirement
Is = I_out * (R+Z)/Rs
I_in = Is + I_out
```

```python
print(f"{I_in=:.2e}, {Is=:.2e}, {I_out=:.2e}")
```

```python
# Power dissipation from P.O.V of I_in
P_dis = I_in** 2 * Z0 # because the input resistance looking into the attenuator is (R + Rs//R) ~ R+Rs ~ 50 ohm
print(f"{P_dis=:.2e}")
```

```python
# Power dissipation from P.O.V of I_out
P_dis_1 = I_in**2 * R + Is**2 * Rs  + I_out**2*R
P_dis_2 = R_eqv * I_out**2
print(f"{P_dis_1=:.2e}\n{P_dis_2=:.2e}")
```
