```python
rho_hdw = 9.928E-9 #ohm-meter
l_4K = 0.3155
l_Still = 0.2755
l_CP = 0.1965
l_MXC = 0.1965

diam = 0.203E-3
area = pi * (diam/2)**2
```

```python
hbar = h/(2*pi)
wq = 2*pi*5E9
T1 = 2E-3
W0 = 2*pi*60E6
```

```python
P_peak = hbar*wq*T1*W0**2/4
P_peak_dBm = 10*np.log10(P_peak/1E-3)
print(f"{P_peak_dBm:.2f} dBm")
```

```python
P_avg = P_peak * np.sqrt(pi)/6
P_avg_dBm = 10*np.log10(P_avg/1E-3)
print(f"{P_avg_dBm:.2f} dBm")
```
